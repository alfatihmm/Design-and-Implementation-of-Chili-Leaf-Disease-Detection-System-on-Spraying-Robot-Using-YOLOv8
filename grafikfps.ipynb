```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca file CSV
df = pd.read_csv('fps_log.csv')

# Ekstrak kolom Time dan FPS
time = df['Time (s)']
fps = df['FPS']

# Plot grafik
plt.plot(time, fps)
plt.title('Grafik FPS terhadap Waktu')
plt.xlabel('Waktu (s)')
plt.ylabel('FPS')
plt.ylim(0, 15)  # Atur rentang sumbu y dari 0 hingga 15
plt.grid(True)
plt.show()
```
