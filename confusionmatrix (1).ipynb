```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Label untuk sumbu x dan y
labels = ["bercak", "berlubang", "buahcabai", "kuning", "background"]

# Matriks kebingungan yang telah dikoreksi
corrected_confusion_matrix = np.array([
    [0.93, 0.03, 0, 0.01, 0.24],
    [0, 0.94, 0, 0.01, 0.26],
    [0, 0, 0.95, 0, 0.03 ],
    [0, 0, 0, 0.91, 0.47],
    [0.07, 0.03, 0.05, 0.06, 0]
])

# Membuat plot dengan matriks kebingungan yang telah dikoreksi
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.1)  # Mengatur ukuran font

# Membuat anotasi tanpa nilai nol
annot = np.where(corrected_confusion_matrix != 0, corrected_confusion_matrix, '')

ax = sns.heatmap(corrected_confusion_matrix, annot=annot, fmt="", cmap="Blues",
                 xticklabels=labels, yticklabels=labels, cbar=True)
ax.set_xlabel('True', fontsize=14)
ax.set_ylabel('Predicted', fontsize=14)
ax.set_title('Confusion Matrix', fontsize=18)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
